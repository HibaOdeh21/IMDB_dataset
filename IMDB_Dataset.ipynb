```python
import pandas as pd 
import matplotlib.pyplot as plt
imdb_path='imdb_top_1000.csv'
imdb_data=pd.read_csv(imdb_path)
print(imdb_data)
```

```python
imdb_data = imdb_data.dropna(axis=0)
print(imdb_data)
```

```python
imdb_data.columns
```

```python
genre_counts=imdb_data['Genre'].value_counts()
g=genre_counts.idxmax()
print(genre_counts)
print("the most popular genre is : ",g)
```

```python
counts=pd.DataFrame(genre_counts).reset_index()
counts.columns=['Genre','Count']
print(counts)
```

```python

plt.figure(figsize=(100,20))
plt.bar(counts['Genre'],counts['Count'])
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Most Popular Genre of shows')
plt.xticks(rotation=90)
plt.show()

```

```python
director_counts=imdb_data['Director'].value_counts()
director_counts_imdb=pd.DataFrame(director_counts).reset_index()
director_counts_imdb.columns=['Director','Count']
d=director_counts.idxmax()
print(director_counts_imdb)
print("the director with the most top-rated movies/TV shows is :",d)
```

```python
plt.figure(figsize=(100,20))
plt.bar(director_counts_imdb['Director'],director_counts_imdb['Count'])
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Top Rated Movies/TV shows by Director')
plt.xticks(rotation=90)
plt.show()
```
